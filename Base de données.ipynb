```python
import sqlalchemy as db

db.__version__
```

```python
from functions import DataBaseV2

# Création d'une base de données
database = DataBaseV2(db_name='test', db_type='sqlite')
```

```python
# Ajout de données provenant d'un fichier csv dans une table
database.insert_csv_to_table('Data', 'data.csv', sep=';')
```

```python
# Affichage des données de la table sous forme de dataframe
database.dataframe('Data')
```

```python
# Création d'un Tableau1
database.create_table('Tableau1', 
                      id_=db.Integer, 
                      colonne1=db.String, 
                      colonne2=db.Integer)
```

```python
# Création d'un Tableau2
database.create_table('Tableau2', id_shop=db.Integer, id_colonne1=db.String, colonne2=db.Integer)
```

```python
# Affichage des tables de la base de données
database.tables
```

```python
# Ajouter une ligne à la base de données
database.insert_row('Tableau2', id_shop=68, id_colonne1='Iphone')
```

```python
# Ajouter plusieurs lignes à la base de données
for i in range(10):
    database.insert_row('Tableau2',id_shop=i, id_colonne1='Iphone'*(i+1), colonne2=99*i)
```

```python
# Afficher le Tableau1
database.select_table('Tableau2')
```

```python
# Supprimer une ligne de la base de données
database.delete_row_by_id('Tableau2', 7)
```

```python
# Afficher le Tableau2
database.dataframe('Tableau2')
```

```python
# Supprimer un Tableau de la base de données
database.delete_table('Tableau1')
```

```python
# Afficher les tables de la base de données
database.tables
```

```python
# Lire un Tableau de la base de données au format sqlalchemy
database.read_table('Tableau2')
```
